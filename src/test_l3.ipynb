### 可視化? with_ pyplot

```python
import numpy as np
import pandas as pd
import scipy as sp

## 実行に無関係なエラーメッセージを消す
import warnings
warnings.filterwarnings('ignore')

from matplotlib import pyplot as plt
%matplotlib inline

## いい感じのグラフを出力する
import seaborn as sns
sns.set()
```

### 折れ線グラフの描画


```python
x = np.arange(1, 16, 1)
y = np.arange(2, 17, 1)
np.random.shuffle(y)
## 描画
plt.plot(x, y, color= 'black')
plt.xlabel("x")
plt.ylabel("random y")

## グラフの画像を保存
plt.savefig("test_l3.jpg")
```

### ヒストグラムの描画

```python
fish_data = np.array([2, 2, 3, 3, 4, 4, 4, 4, 5, 5, 6, 7])
sns.distplot(fish_data, bins=5, color= 'black')
```

### 2変量データに対するヒストグラム

```python
fish_multi = pd.read_csv('data/3-3-2-fish_multi_2.csv')
fish_multi.head()
```

```python
fish_multi.groupby("species").describe()
```

```python
length_a = fish_multi.query('species == "A"')["length"]
length_b = fish_multi.query('species == "B"')["length"]
```

```python
sns.distplot(length_a, kde=False, bins = 5)
sns.distplot(length_b, kde=False, bins = 5)
```

### 箱ひげ図の描画

```python
sns.boxplot(x="species", y="length", data=fish_multi)
```
