```python
x = np.arange(1, 16, 1)
y = np.arange(2, 17, 1)
np.random.shuffle(y)
## 描画
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("random y")

## グラフの画像を保存
plt.savefig("test_l3.png")
```
